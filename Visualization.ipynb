## Define

### Price:

\begin{gather*}
\frac{Highest(Price_{cmponents}) \times min_{bid}(Highest(Price_{cmponents}))}{1000}
\end{gather*}

### Auction:

\begin{gather*}
Auction(\frac{price_{NLT}}{price_{Component}}\times 1000) \iff Price_{NLT} > Price_{Component} 
\end{gather*}

### Redeem:

\begin{gather*}
Redeem(\frac{price_{NLT}}{price_{Component}}) \iff Price_{NLT} < Price_{Component}
\end{gather*}
